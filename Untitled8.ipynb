```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp = pd.read_csv ('finaldataset.csv')
```

```python
temp
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.regplot(data=temp, x='TemperatureC', y='Humidity level', ci=95, line_kws={'color': 'red'})
plt.xlabel('TemperatureC')
plt.ylabel('Humidity level')
plt.title('Scatter Plot with Loess Regression Line and Confidence Bands')
plt.legend(['Regression Line (95% CI)'])
plt.show()
```

```python
import matplotlib.pyplot as plt
# Sample data
x_values = [5,10,15,20,25]
y_values = [50,60,70,80,90]
# Plotting the line chart
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='Line Chart')
# Adding labels and title
plt.xlabel('TemperatureC')
plt.ylabel('Sun rise')
plt.title('Simple Line Chart')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting the histogram
plt.hist(temp, bins=20, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('TemperatureC')
plt.ylabel('Sun rise')
plt.title('Histogram Example')

# Display the plot
plt.show()
```
